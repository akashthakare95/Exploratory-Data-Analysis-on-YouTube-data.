### 13. Does the number of views have any effect on how many people disliked the video? Support your
### answer with a metric and a plot.

````python
ds['view_count'].corr(ds['dislikes'])
````

````python
import matplotlib.pyplot as plt
````

````python
fig, ax = plt.subplots(figsize=(10,4))
ax.scatter(ds['view_count'], ds['dislikes'])
ax.set_xlabel('Views Of Videos')
ax.set_ylabel('dislikes Of Videos')
plt.title("Relation between views and dislikes")
plt.show()
````

```
As we can see from the correlation between views and dislikes on YouTube videos is (0.6844687753905542).This means that, on average, videos that have more views also tend to have more dislikes, and vice versa.

Looking at the plot above, we can see most of data points are lying on the bottom left side but there are points which are far from the populated data like top right corner. Those are potential outliers.
```
